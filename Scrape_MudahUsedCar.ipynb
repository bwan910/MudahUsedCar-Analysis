```python
# These empty lists are needed to store each categories of data

car_link = []
car_name = []
car_price = []
car_condition = []
car_year = []
car_mileage = []
car_capacity = []
```

```python
for page in range(3, 240):

    url = "https://www.mudah.my/malaysia/cars-for-sale?o={}".format(page)

    headers = {'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/86.0.4240.198 Safari/537.36'}

    # create variable to store webpage with pagenumber
    response = requests.get(str(url), headers= headers)
    #print(response.status_code)
    # object to scrape html elements
    soup = BeautifulSoup(response.text, "html.parser")
    time.sleep(5)

    #print(soup)

    soup_content = soup.findAll("div", {"class": "sc-cmIlrE frzaNp"})
    #print(len(soup_content))

    for x in soup_content:
        link = x.find("a", {"class": "sc-gleUXh gUqogd"})['href']    
        print(link)
        car_link.append(link)

        name = x.find("a", {"class": "sc-gleUXh gUqogd"}).text
        print(name)
        car_name.append(name)

        price =  x.find("div", {"class": "sc-bNQFlB dNipmi"}).text
        print(price)
        car_price.append(price)
        
        condition = x.find("div",{"title":"Condition"}).get_text()
        print(condition)
        car_condition.append(condition)
        
        
        year = x.find("div",{"title":"Manufactured Year"})
        clean = re.compile('<.*?>')
        year = re.sub(clean, '', str(year))
        print(year)
        car_year.append(year)
        
        mileage = x.find("div",{"title":"Mileage"}).get_text()
        print(mileage)
        car_mileage.append(mileage)
        
        
        capacity = x.find("div",{"title":"Engine capacity"}).get_text()
        print(capacity)
        car_capacity.append(capacity)
        
        
        #div = x.find("div", {"class": "sc-iQNlJl kRXswG"})
        #area = x.find("span", {"class": "sc-iuJeZd eYhUll"})
        #print(area)
        
        
        
        

```

```python
# Create a dictionary to store all data according to category

data = {
   'Car link': car_link, 
    'Car name': car_name, 
    'Car price': car_price, 
    'Car condition': car_condition, 
    'Car year': car_year,
    'Car mileage': car_mileage, 
    'Car cpacity':car_capacity
}

print(data)

    
```

## Store into a Dataframe

```python
df = DataFrame(data, columns=[
    'Car link',
    'Car name',
    'Car price',
    'Car condition',
    'Car year',
    'Car mileage',
    'Car cpacity'
])

df
```
